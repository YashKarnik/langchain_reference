```python
from langchain_core.prompts import ChatPromptTemplate
from langchain.chains.retrieval import create_retrieval_chain
from langchain.chains.combine_documents import create_stuff_documents_chain

from langchain_openai import AzureChatOpenAI

os.environ["AZURE_OPENAI_API_KEY"] = "##"
os.environ["AZURE_OPENAI_ENDPOINT"] = "https://gptembedder.openai.azure.com/"

template = ChatPromptTemplate.from_template("""
    You are an AI assistant. Your job is to summarize the below topic.
    Topic: {context}
""")


base_chat_llm = AzureChatOpenAI(
    azure_deployment="gpt-4o",  # or your deployment
    api_version="2024-12-01-preview",  # or your api version
)

document_chain = create_stuff_documents_chain(base_chat_llm, template)
retrieval_chain = create_retrieval_chain(vector_store2.as_retriever(), document_chain)
```

```python
content = retrieval_chain.invoke({"input": "Encoders"})
```
